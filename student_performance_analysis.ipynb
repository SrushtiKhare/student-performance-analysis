```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

print(df.head())
print(df.describe())

# Create average score
df["average_score"] = (df["math score"] + df["reading score"] + df["writing score"]) / 3

# Graph 1
plt.figure()
df.groupby("gender")["average_score"].mean().plot(kind="bar")
plt.title("Average Score by Gender")
plt.ylabel("Score")
plt.show()

# Graph 2
plt.figure()
df.groupby("parental level of education")["average_score"].mean().plot(kind="bar")
plt.xticks(rotation=45)
plt.title("Parental Education vs Average Score")
plt.show()

# Graph 3
plt.figure()
df.groupby("lunch")["average_score"].mean().plot(kind="bar")
plt.title("Lunch Type vs Score")
plt.show()
```
